## Treating null values

```python
#removing records with missing target values
df = df[df["Life expectancy"].notnull()]
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# percentage missing values
print((df.isnull().sum()/2928)*100)
```

```python
#removing records with less than 2% missing values
# for i in df.columns:
#     if df[i].isnull().sum() < 33:
#         df = df[df[i].notnull()]
# df.shape
```

```python
null_columns = list(df.columns[df.isnull().any()])
```

```python
null_columns
```

```python
#finding missing values grouped by countries
df[null_columns] = df.groupby('Country')[null_columns].transform(lambda x:x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
null_columns = list(df.columns[df.isnull().any()])
```

```python
null_columns
```

```python
#replacing the null values in remaining columns with their respective means
for i in null_columns:
    df[i]=df[i].fillna(df[i].mean())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#dropping Country and Year columns
df = df.drop(columns = ['Country','Year'])
df.head()
```

```python
df.Status.value_counts()
```

```python
#mapping Status column to numerical values
df.Status = df.Status.map({'Developing':0,'Developed':1})
df.head()
```

## Model building and evaluation

```python
#Splitting train and test data
from sklearn.model_selection import train_test_split
df_train,df_test = train_test_split(df,train_size=0.7,test_size=0.3,random_state=100)
```

```python
#Scaling training data
from sklearn.preprocessing import StandardScaler

#creating a scaling object
scaler = StandardScaler()

varlist = list(df.columns)

#scaling using 'fit_transform'
df_train[varlist]=scaler.fit_transform(df_train[varlist])
```

```python
df_train.head()
```

```python
#Split train data into x and y
y_train=df_train.pop('Life expectancy')
x_train= df_train
```

```python
from sklearn.linear_model import LinearRegression
#creating a linear regression object
lm = LinearRegression()
```

## Model building using RFE

```python
from sklearn.feature_selection import RFE

#RFE with 15 features
rfe1 = RFE(lm,15)

#fit with 15 features
rfe1.fit(x_train, y_train)

#print the boolean result
print(rfe1.support_)
```

```python
list(zip(x_train.columns,rfe1.support_,rfe1.ranking_))
```

```python
import statsmodels.api as sm

#subset the features selected by rfe1
col1=x_train.columns[rfe1.support_]

#subsetting training data from 15 selected columns
x_train_rfe1=x_train[col1]

#add a constant to the model
x_train_rfe1 = sm.add_constant(x_train_rfe1)
x_train_rfe1.head()
```

```python
#Fitting the model with 15 variables
lm1 = sm.OLS(y_train,x_train_rfe1).fit()
print(lm1.summary())
```

### The value of p for 'percentage expenditure' and 'Measles' is coming out to be greater than the desirable value of 0.05

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#checking VIF values of feature variables
vif=pd.DataFrame()
vif['Features'] = x_train_rfe1.columns
vif['VIF']=[variance_inflation_factor(x_train_rfe1.values,i) for i in range(x_train_rfe1.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

### The VIF values of 'under-five deaths', 	'infant deaths', 'GDP' columns is coming out to be way greater than the desirable value of 5.

```python
#RFE with 10 features
rfe2 = RFE(lm,10)

#fit with 10 features
rfe2.fit(x_train, y_train)
```

```python
#subset the features selected by rfe1
col2=x_train.columns[rfe2.support_]

#subsetting training data from 15 selected columns
x_train_rfe2=x_train[col2]

#add a constant to the model
x_train_rfe2 = sm.add_constant(x_train_rfe2)
x_train_rfe2.head()
```

```python
#Fitting the model with 10 variables
lm2 = sm.OLS(y_train,x_train_rfe2).fit()
print(lm2.summary())
```

```python
#checking VIF values of feature variables
vif2=pd.DataFrame()
vif2['Features'] = x_train_rfe2.columns
vif2['VIF']=[variance_inflation_factor(x_train_rfe2.values,i) for i in range(x_train_rfe2.shape[1])]
vif2['VIF']=round(vif2['VIF'],2)
vif2 = vif2.sort_values(by='VIF', ascending=False)
vif2
```

```python
#dropping 'under-five deaths','infant deaths' columns as their VIF values are greater than 5
x_train_rfe3 = x_train_rfe2.drop(columns=['under-five deaths','infant deaths'])
```

```python
#Fitting the model with 10 variables
lm3 = sm.OLS(y_train,x_train_rfe3).fit()
print(lm3.summary())
```

```python
#checking VIF values of feature variables
vif3=pd.DataFrame()
vif3['Features'] = x_train_rfe3.columns
vif3['VIF']=[variance_inflation_factor(x_train_rfe3.values,i) for i in range(x_train_rfe3.shape[1])]
vif3['VIF']=round(vif3['VIF'],2)
vif3 = vif3.sort_values(by='VIF', ascending=False)
vif3
```

All the VIF values and p-values seem to be in a good range. Also the adjuster R-squared value has dropped from 0.824 with 15 variables to just 0.807 using 8 vaiables. So let's proceed with this model.

## Residual Analysis

```python
y_train_Le = lm3.predict(x_train_rfe3)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_Le), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)   
```

Error terms seem to be normally distributed

```python
#plotting y_test vs y_test_pred to understand the spread
fig = plt.figure()
plt.scatter(y_test,y_test_pred)
fig.suptitle('y_test vs y_test_pred', fontsize = 20)
plt.xlabel('y_test', fontsize=18)                       
plt.ylabel('y_test_pred', fontsize=16) 
```
